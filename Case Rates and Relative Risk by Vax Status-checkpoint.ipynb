# Covid Case Rates in Ontario by Vax Status:
# August 18th, 2021 to March 10th, 2022

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly as py
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.patches as mpatches
```

```python
# Getting cases by vax status and cases by vax status *rates*

data = pd.read_csv('https://data.ontario.ca/dataset/752ce2b7-c15a-4965-a3dc-397bf405e7cc/resource/eed63cf2-83dd-4598-b337-b288c0a89a16/download/cases_by_vac_status.csv')
df = data.copy()
```

```python
df.head(3)
```

```python
# I just want to see when they started reporting on case *rates* by vax status (7-day average)

ticker = []
for x in df['cases_unvac_rate_7ma']:
    if np.isnan(x) == True:
        ticker += [1]
        
start = sum(ticker)

# Just showing that yeah they started tracking case rate by vax status at the 9th index location

df[start-1:start+2]
```

```python
n = 15 + len(df[start:]) - len(df[start:214])

fig= plt.figure(figsize=(30,10))
axes= fig.add_axes([1,1,1,1])
axes.tick_params(axis='both', which='major', width=3, length=9)
axes.set_facecolor('#f3fcfa')
axes.grid(axis='y', c = '#03100c', alpha=0.3, linestyle=':')

axes.plot(df[start:214]['Date'],df[start:214]['cases_unvac_rate_7ma'],  c = '#8123a0', lw = 10, alpha = 0.8, zorder=1) 
axes.plot(df[start:214]['Date'],df[start:214]['cases_full_vac_rate_7ma'],  c = '#23a081', lw = 10, alpha = 0.8,zorder=2) 
axes.plot(df[start:214]['Date'],df[start:214]['cases_partial_vac_rate_7ma'],  c = '#F78DA7', lw = 10, alpha = 0.2,zorder=0) 

plt.yticks(fontsize = 30)
plt.xticks(ticks=[23-n, 52-n, 84-n, 114-n, 145-n, 176-n, 204-n], labels=['Sept 2021', 'Oct 2021', 'Nov 2021', 'Dec 2021', 'Jan 2022', 'Feb 2022', 'March 2022'], fontsize=28)

axes.legend(['Unvaxxed','Fully Vaxxed', 'Partially Vaxxed'], fontsize=31, shadow=True, edgecolor = '#3d989b')

plt.scatter(137-n,32, s=350,zorder=3, c = '#d43b61')
plt.scatter(171-n,34.7, s=350,zorder=3, c = '#d43b61')
plt.scatter(200-n,11.55, s=350,zorder=3, c = '#d43b61')


plt.ylabel('Rate per 100,000 (7-Day Average)', labelpad = 10, fontsize = 35)
plt.title('Case Rates by Vax Status, Ontario', size = 50, pad = 40)



# I should add horizontal grid lines behind ZORDER the curvers
# also need to do a risk graph unv'd versus fully v'd
```

## Comparative risk of infection: vaxxed vs unvaxxed

```python
risk = df[start:214]['cases_unvac_rate_7ma'] / df[start:214]['cases_full_vac_rate_7ma']
risk = pd.DataFrame(risk)
risk.columns = ['Risk Ratio']
date = df[start:214]['Date']
risk['Date'] = date
```

```python
risk.head(3)
```

```python
n = 15 + len(df[start:]) - len(df[start:214])

fig= plt.figure(figsize=(30,10))
axes= fig.add_axes([1,1,1,1])
fig.patch.set_facecolor('#F9F9F9')

axes.add_patch(mpatches.Rectangle((-8.8,1.1),width = 73, height = 3.8, ec = '#555555', fill = True, lw = 0, fc ='#F9F9F9', alpha = 0.6))

# Vax Pass starts
axes.annotate('', arrowprops = { 'arrowstyle' : 'fancy', 'connectionstyle' : 'arc3,rad=-0.1', 'color' : '#23a081', 'alpha':0.8}, xy = (45-n,6.075145), xytext = (40,6.4),zorder=0, size=32)
axes.annotate('Vax Pass starts', xy = (29,6.6), size=34, alpha = 0.8)


# Overall trend
axes.annotate('', arrowprops = { 'arrowstyle' : 'fancy', 'connectionstyle' : 'arc3,rad=0.1', 'color' : '#de425b', 'alpha':0.4}, xy = (100-n,3.5), xytext = (-5,6.5),zorder=1, size=32)
axes.annotate('Prior to Omicron \narriving there is a clear \ndownward trend in the relative \nrisk of unvaxxed people contracting Covid\ncompared to fully vaxxed people.\n\nPotentially due to waning vaccine efficacy\nas well as more unvaxxed people acquiring\nnatural immunity?', xy = (-6,1.5), size=32, alpha = 0.65)


# Omicron first identified in Ontario
axes.annotate('', arrowprops = { 'arrowstyle' : 'fancy', 'connectionstyle' : 'arc3,rad=0.1', 'color' : '#23a081', 'alpha':0.8}, xy = (110-n,4.172857), xytext = (70,5.3),zorder=0, size=32)
axes.annotate('Omicron first\ndetected in Ontario', xy = (55,5.5), size=34, alpha = 0.8)


# Omicron majority of cases
axes.annotate('', arrowprops = { 'arrowstyle' : 'fancy', 'connectionstyle' : 'arc3,rad=-0.1', 'color' : '#23a081', 'alpha':0.8}, xy = (126.5-n,3.1), xytext = (132-n,6.2),zorder=0, size=32)
axes.annotate('Public Health Ontario finds\nthat the majority of Covid \ncases are Omicron', xy = (110-n,6.3), size=29, alpha = 0.8)


# Risk Ratio bottoms out
axes.annotate('', arrowprops = { 'arrowstyle' : 'fancy', 'connectionstyle' : 'arc3,rad=0.1', 'color' : '#23a081', 'alpha':0.8}, xy = (154.5-n,0.9), xytext = (150-n,3),zorder=0, size=32)
axes.annotate('Unvaxxed risk ratio\nbottoms out at 0.74', xy = (135-n,3.1), size=32, alpha = 0.8)


# Highest risk in 2022
axes.annotate('', arrowprops = { 'arrowstyle' : 'fancy', 'connectionstyle' : 'arc3,rad=-0.1', 'color' : '#23a081', 'alpha':0.8}, xy = (183-n,1.3), xytext = (175-n,4.2),zorder=0, size=32)
axes.annotate('Unvaxxed risk ratio hits\nits 2022 peak at 1.18', xy = (160-n,4.35), size=32, alpha = 0.8)


# Ontario stops reporting unvaxxed case rates
axes.annotate('', arrowprops = { 'arrowstyle' : 'fancy', 'connectionstyle' : 'arc3,rad=-0.1', 'color' : '#23a081', 'alpha':0.8}, xy = (213-n,0.9), xytext = (210-n,2.1),zorder=0, size=32)
axes.annotate('Ontario stops reporting\nunvaxxed case rates', xy = (188-n,2.3), size=28, alpha = 0.8)


axes.tick_params(axis='both', which='major', width=3, length=9)
axes.set_facecolor('#f3fcfa')
axes.grid(axis='y', c = '#03100c', alpha=0.3, linestyle=':')

axes.plot(risk['Date'],risk['Risk Ratio'],  c = '#8123a0', lw = 12, alpha = 0.95, zorder=1) 

plt.yticks(fontsize = 30)
plt.xticks(ticks=[23-n, 52-n, 84-n, 114-n, 145-n, 176-n, 204-n], labels=['Sept 2021', 'Oct 2021', 'Nov 2021', 'Dec 2021', 'Jan 2022', 'Feb 2022', 'March 2022'], fontsize=28)

plt.ylabel('Risk Ratio (X times as likely)', labelpad = 10, fontsize = 35)
plt.title('Relative Risk of Testing + for Covid, Unvaxxed vs Fully Vaxxed (Ontario)', size = 50, pad = 25)

axes.legend(['Relative Risk of Infection for Unvaxxed'], fontsize=31, shadow=True, edgecolor = '#3d989b')



plt.scatter(126-n,2.9747, s=350,zorder=3, c = '#d43b61', alpha = 0.8)
plt.scatter(111-n,4.142857, s=350,zorder=3, c = '#d43b61', alpha = 0.8)
plt.scatter(183-n,1.187238, s=350,zorder=3, c = '#d43b61', alpha = 0.8)
plt.scatter(155-n,0.740097, s=350,zorder=3, c = '#d43b61', alpha = 0.8)
plt.scatter(44-n,6.075145, s=350,zorder=3, c = '#d43b61', alpha = 0.8)
plt.scatter(213-n,0.754141, s=350,zorder=3, c = '#d43b61', alpha = 0.8)

fig.savefig('annotated risk ratio plot.png', bbox_inches='tight', dpi=250)
```
